```python
import numpy
import statistics
import math
import matplotlib.pyplot as plt 

age = numpy.array([23, 23, 27, 27, 39, 41, 47, 49, 50, 52, 54, 54, 56, 57, 58, 58, 60, 61])
mg = numpy.array([9.5, 26.5, 7.8, 17.8, 31.4, 25.9, 27.4, 27.2, 31.2, 34.6, 42.5, 28.8, 33.4, 30.2, 34.1, 32.9, 41.2, 35.7])

print("Moyenne age :")
print(age.mean())

print("Moyenne mg :")
print(mg.mean())

print("Mediane age:")
print(numpy.median(age, 0))

print("Mediane mg:")
print(numpy.median(mg, 0))

print("Ecart type age :")
print(numpy.nanstd(age))

plt.plot(age, mg) 
# naming the x axis 
plt.xlabel('x - age') 
# naming the y axis 
plt.ylabel('y - mg') 

plt.plot(age[0],mg[0],'ro') 
plt.plot(age[2],mg[2],'ro') 

print("Distance euclidienne :")
print((age[0] - age[2])**2 + (mg[0] - mg[2])**2 )
# data = numpy.array([[1.5,1.7],[2,1.9],[1.6,1.8],[1.2,1.5],[1.5,1]])
```
